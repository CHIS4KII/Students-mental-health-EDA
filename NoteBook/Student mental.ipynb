```python
gender_counts = df['gender'].value_counts()

plt.figure(figsize=(4, 4))
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=90, colors=['cyan', 'pink'], explode=(0, 0.1))
plt.title('Gender Distribution')
plt.show()
```

```python
male_values = df[df['gender'] == 'Male'][['depression', 'anxiety', 'isolation' , 'future_insecurity']]
female_values = df[df['gender'] == 'Female'][['depression', 'anxiety', 'isolation' , 'future_insecurity']]

# Creating a figure with 4 subplots, 2 rows and 2 columns
plt.figure(figsize=(10, 8))

# Depression
plt.subplot(2, 2, 1)
plt.plot(male_values['depression'].values, label='Male', color='blue')
plt.plot(female_values['depression'].values, label='Female', color='orange')
plt.title('Depression')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Anxiety
plt.subplot(2, 2, 2)
plt.plot(male_values['anxiety'].values, label='Male', color='blue')
plt.plot(female_values['anxiety'].values, label='Female', color='orange')
plt.title('Anxiety')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Isolation
plt.subplot(2, 2, 3)
plt.plot(male_values['isolation'].values, label='Male', color='blue')
plt.plot(female_values['isolation'].values, label='Female', color='orange')
plt.title('Isolation')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Future Insecurity
plt.subplot(2, 2, 4)
plt.plot(male_values['future_insecurity'].values, label='Male', color='blue')
plt.plot(female_values['future_insecurity'].values, label='Female', color='orange')
plt.title('Future Insecurity')
plt.xlabel('Index')
plt.ylabel('Values')
plt.legend()

# Adjust layout for better spacing
plt.tight_layout()

# Show the plot
plt.show()
```

```python
df['academic_year'] = df['academic_year'].str.replace('st' , '').str.replace('rd' , '').str.replace('nd' , '').str.replace('th' , '').str.replace('year' , '')
df['cgpa'] = df['cgpa'].str.split('-').str[0]
df
```

```python
university_df = df.groupby('university')[['depression', 'anxiety', 'isolation', 'future_insecurity']].agg(['sum', 'mean', 'min', 'max'])
university_df
```

```python
plt.figure(figsize = (10 , 6))
depression_mean = university_df.xs('mean', level=1, axis=1)['depression']
plt.figure(figsize=(10, 6))
plt.bar(university_df.index, depression_mean , color = 'black')
plt.xlabel('Universities')
plt.ylabel('Mean of Depression Responses')
plt.title('Depression Levels Across Universities')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10 , 6))
anxiety_meean = university_df.xs('mean', level=1, axis=1)['anxiety']
plt.figure(figsize=(10, 6))
plt.bar(university_df.index, anxiety_meean , color = 'black')
plt.xlabel('Universities')
plt.ylabel('Mean of Anxiety Responses')
plt.title('Anxiety Levels Across Universities')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10 , 6))
isolation_meean = university_df.xs('mean', level=1, axis=1)['isolation']
plt.figure(figsize=(10, 6))
plt.bar(university_df.index, isolation_meean , color = 'black')
plt.xlabel('Universities')
plt.ylabel('Mean of Isolation Responses')
plt.title('Isolation Levels Across Universities')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10 , 6))
future_insecurity_meean = university_df.xs('mean', level=1, axis=1)['future_insecurity']
plt.figure(figsize=(10, 6))
plt.bar(university_df.index, future_insecurity_meean , color ='black')
plt.xlabel('Universities')
plt.ylabel('Mean of Future Insecurity Responses')
plt.title('Future Insecurity Levels Across Universities')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['sports_engagement'] = df['sports_engagement'].str.replace('times' , '').str.replace('+' , '').str.split('-').str[0]
```

```python
life_style_factor = ['sports_engagement' , 'campus_discrimination' , 'residential_status' , 'depression' , 'anxiety' , 'isolation']
df[life_style_factor]
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.histplot(data=df, x="depression", hue="residential_status", kde=True, ax=axes[0, 0])
axes[0, 0].set_title("Depression by Residential Status")
sns.histplot(data=df, x="anxiety", hue="residential_status", kde=True, ax=axes[0, 1])
axes[0, 1].set_title("Anxiety by Residential Status")
sns.histplot(data=df, x="isolation", hue="residential_status", kde=True, ax=axes[1, 0])
axes[1, 0].set_title("Isolation by Residential Status")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.violinplot(data=df, x="sports_engagement", y="depression", ax=axes[0, 0])
axes[0, 0].set_title("Depression by Sports Engagement")
sns.violinplot(data=df, x="sports_engagement", y="anxiety", ax=axes[0, 1])
axes[0, 1].set_title("Anxiety by Sports Engagement")
sns.violinplot(data=df, x="sports_engagement", y="isolation", ax=axes[1, 0])
axes[1, 0].set_title("Isolation by Sports Engagement")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
sns.boxplot(data=df, x="campus_discrimination", y="depression", ax=axes[0, 0])
axes[0, 0].set_title("Depression by Campus Discrimination")
sns.boxplot(data=df, x="campus_discrimination", y="anxiety", ax=axes[0, 1])
axes[0, 1].set_title("Anxiety by Campus Discrimination")
sns.boxplot(data=df, x="campus_discrimination", y="isolation", ax=axes[1, 0])
axes[1, 0].set_title("Isolation by Campus Discrimination")
axes[1, 1].axis("off")

plt.tight_layout()
plt.show()
```
